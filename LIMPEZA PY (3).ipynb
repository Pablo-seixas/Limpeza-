```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

# Conectar ao banco de dados SQLite
conn = sqlite3.connect('dados_limpeza.db')

# Consultar dados
query = 'SELECT * FROM limpeza'
df_sqlite = pd.read_sql_query(query, conn)

# Fechar a conexão
conn.close()

# Criar um grafo direcionado
G = nx.DiGraph()

# Adicionar nós (Nomes) ao grafo
for nome in df_sqlite['Nome']:
    G.add_node(nome)

# Adicionar arestas (conexões entre Nomes)
for i in range(len(df_sqlite)-1):
    G.add_edge(df_sqlite['Nome'][i], df_sqlite['Nome'][i+1])

# Desenhar o grafo
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, font_weight='bold', node_size=2000, node_color='skyblue', font_size=10, edge_color='gray', width=1.5, alpha=0.7)

plt.title('Grafo de Relacionamento entre Nomes')
plt.show()
```

```python
import pandas as pd
from scipy import stats

# Supondo que 'df' seja o seu DataFrame
df = pd.DataFrame({
    'coluna1': [1, 2, None, 4, 5],
    'coluna2': [10, 20, 30, 40, None],
    'coluna3': ['A', 'B', 'C', None, 'E']
})

# 1. Preencher valores ausentes com uma constante
df = df.fillna(0)  # Substitua 0 pelo valor desejado

# 2. Identificar e remover outliers usando z-score
numeric_cols = ['coluna1', 'coluna2']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df = df.dropna(subset=numeric_cols)

z_scores = stats.zscore(df[numeric_cols])
df = df[(z_scores < 3) & (z_scores > -3)]

# 3. Converter tipos de dados se necessário
df['coluna1'] = pd.to_numeric(df['coluna1'], errors='coerce')

# ... (outras etapas de limpeza conforme necessário)

# 7. Continuar análises ou salvar o DataFrame limpo
df.to_csv('dados_limpos.csv', index=False)
```
