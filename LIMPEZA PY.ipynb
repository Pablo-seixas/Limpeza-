```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Obtenha o caminho do desktop
desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

# Caminho completo para o arquivo limpeza.xlsx no desktop
file_path = os.path.join(desktop_path, "limpeza.xlsx")

# Carregue os dados do arquivo Excel
df = pd.read_excel(file_path)

# Supondo que 'Nome,Sexo,Telefone,CPF,Idade,Empresa,Horas_Trabalhadas' seja o nome da coluna
# Divida a coluna em várias colunas
df[['Nome', 'Sexo', 'Telefone', 'CPF', 'Idade', 'Empresa', 'Horas_Trabalhadas']] = df['Nome,Sexo,Telefone,CPF,Idade,Empresa,Horas_Trabalhadas'].str.split(',', expand=True)

# Converta a coluna 'Idade' para o tipo correto (int ou float)
df['Idade'] = df['Idade'].astype(int)

# Verificar os tipos de dados
print(df.dtypes)

# Estatísticas descritivas
print(df.describe())

# Criar gráfico de barras empilhadas
df.groupby(['Idade', 'Sexo']).size().unstack().plot(kind='bar', stacked=True, title='Distribuição por Sexo e Idade')
plt.xlabel('Idade')
plt.ylabel('Número de Pessoas')
plt.show()
```

```python

# Obtenha o caminho do desktop
desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

# Caminho completo para o arquivo limpeza.xlsx no desktop
file_path = os.path.join(desktop_path, "limpeza.xlsx")

# Carregue os dados do arquivo Excel
df = pd.read_excel(file_path)

# Supondo que 'Nome,Sexo,Telefone,CPF,Idade,Empresa,Horas_Trabalhadas' seja o nome da coluna
# Divida a coluna em várias colunas
df[['Nome', 'Sexo', 'Telefone', 'CPF', 'Idade', 'Empresa', 'Horas_Trabalhadas']] = df['Nome,Sexo,Telefone,CPF,Idade,Empresa,Horas_Trabalhadas'].str.split(',', expand=True)

# Converta a coluna 'Idade' para o tipo correto (int ou float)
df['Idade'] = df['Idade'].astype(int)

# Mapeamento de cores para cada sexo
cores = {'Homem': 'blue', 'Mulher': 'magenta'}

# Criar gráfico de dispersão
plt.scatter(df['Idade'], df['Horas_Trabalhadas'], c=df['Sexo'].map(cores), alpha=0.5)
plt.title('Distribuição de Horas Trabalhadas por Idade e Sexo')
plt.xlabel('Idade')
plt.ylabel('Horas Trabalhadas')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', label='Homem', markerfacecolor='blue', markersize=10),
                    plt.Line2D([0], [0], marker='o', color='w', label='Mulher', markerfacecolor='magenta', markersize=10)],
           title='Sexo')
plt.grid(True)
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd

# Supondo que 'df' seja o seu DataFrame
df = pd.DataFrame({
    'Nome': ['Maria Silva', 'João Santos', 'Ana Oliveira', 'Pedro Costa', 'Carla Pereira',
             'Luiz Souza', 'Fernanda Santos', 'Rafael Lima', 'Mariana Costa', 'Gustavo Oliveira',
             'Juliana Pereira', 'André Souza', 'Camila Santos', 'Lucas Lima', 'Isabela Costa'],
    'Sexo': ['Mulher', 'Homem', 'Mulher', 'Homem', 'Mulher', 'Homem', 'Mulher', 'Homem', 'Mulher', 'Homem',
             'Mulher', 'Homem', 'Mulher', 'Homem', 'Mulher'],
    'Telefone': [1122334455, 5544332211, 9988776655, 1122113344, 6677889900, 1122334455, 5544332211, 9988776655,
                  1122113344, 6677889900, 1122334455, 5544332211, 9988776655, 1122113344, 6677889900],
    'CPF': ['123.456.789-09', '987.654.321-09', '234.567.890-12', '543.210.987-76', '876.543.210-98',
            '345.678.901-23', '012.345.678-90', '987.654.321-09', '234.567.890-12', '543.210.987-76',
            '876.543.210-98', '345.678.901-23', '012.345.678-90', '987.654.321-09', '234.567.890-12'],
    'Idade': [30, 25, 28, 32, 35, 29, 27, 31, 33, 26, 29, 34, 30, 28, 32],
    'Empresa': ['Empresa de Limpeza A', 'Empresa de Limpeza B', 'Empresa de Limpeza C', 'Empresa de Limpeza A',
                'Empresa de Limpeza B', 'Empresa de Limpeza C', 'Empresa de Limpeza A', 'Empresa de Limpeza B',
                'Empresa de Limpeza C', 'Empresa de Limpeza A', 'Empresa de Limpeza B', 'Empresa de Limpeza C',
                'Empresa de Limpeza A', 'Empresa de Limpeza B', 'Empresa de Limpeza C'],
    'Horas_Trabalhadas': [40, 35, 38, 42, 37, 40, 35, 38, 41, 37, 40, 36, 39, 34, 43]
})

# Criar uma figura
fig, ax = plt.subplots(figsize=(12, 8))

# Adicionar cores diferentes para representar diferentes empresas
colors = {'Empresa de Limpeza A': 'red', 'Empresa de Limpeza B': 'blue', 'Empresa de Limpeza C': 'green'}
df['Cor'] = df['Empresa'].map(colors)

# Criar o gráfico de dispersão
scatter = ax.scatter(df['Idade'], df['Horas_Trabalhadas'], c=df['Cor'], s=100, alpha=0.7)

# Adicionar rótulos e título
plt.title('Relação entre Idade e Horas Trabalhadas por Pessoa')
plt.xlabel('Idade')
plt.ylabel('Horas Trabalhadas')

# Adicionar legenda para empresas
legend_labels = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[empresa], markersize=10, label=empresa) for empresa in colors]
ax.legend(handles=legend_labels, title='Empresa')

# Função para atualizar os gráficos a cada quadro
def update(frame):
    # Atualizar os dados do gráfico de dispersão
    scatter.set_offsets(df[['Idade', 'Horas_Trabalhadas']].iloc[:frame+1].values)

# Criar a animação
animation = FuncAnimation(fig, update, frames=len(df), interval=1000, repeat=False)

# Exibir a animação
plt.show()

# Diretório de destino
destination_directory = r'X:\pablo dev\anaconda\lib\site-packages\matplotlib'

# Salvar a animação no diretório especificado usando o Pillow
animation_file = os.path.join(destination_directory, 'animated_scatter.gif')
animation.save(animation_file, writer='pillow', fps=1)
```
